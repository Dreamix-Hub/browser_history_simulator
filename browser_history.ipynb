```python
import requests
from bs4 import BeautifulSoup
```

```python
class Node:
    def __init__(self,url):
        self.prev = None
        self.url = url
        self.next = None

class BrowserHistory:
    def __init__(self):
        self.head = None
        self.current = None
        self.tail = None
```

```python
def visit_website(self, url):
    
    url = "https://www." + url
    new_node = Node(url)
    
    
    try:
        response = requests.get(url, timeout=5)
        if response.status_code == 200:
            soup = BeautifulSoup(response.text, "html.parser")
            title = soup.title.string if soup.title else "No title!"
            print("Page title:",title)
    except requests.exceptions.RequestException:
        print("Network error!")
        
    # first visit
    if self.head is None:
        self.head = self.current = self.tail = new_node
        return
    
    # moving to the latest page (we went back before)
    # when we go back and visit new site this will automatically remove the next page with the new page
    if self.current.next is not None:
        self.current.next = None
        self.tail = self.current
            
    # add new node
    self.tail.next = new_node
    new_node.prev = self.tail
    self.tail = new_node
    self.current = new_node # current page
    return

BrowserHistory.visit_website = visit_website
```

```python
def go_back(self):
    
    if self.current is None:
        print("No history!")
        return
    
    if self.current.prev is None:
        print("No page previous")
        return
    
    self.current = self.current.prev
    return self.current.url
    
    
BrowserHistory.go_back = go_back
```

```python
def go_forward(self):
    if self.current is None:
        print("No history")
        return
    
    if self.current.next is None:
        print("No page next")
        return
    
    self.current = self.current.next
    return self.current.url

BrowserHistory.go_forward = go_forward
```

```python
def __str__(self):
    if self.head is None:
        print("No History!")
        return
    ret_str = ""
    temp = self.head
    while temp is not None:
        ret_str += temp.url + " --> "
        temp = temp.next
    return ret_str.strip(' --> ')
BrowserHistory.__str__ = __str__
```

```python
browser = BrowserHistory()

browser.visit_website('google.com')
browser.visit_website('youtube.com')
browser.visit_website("github.com")
print(browser)


browser.go_back()
print(browser)
browser.go_forward()
print(browser)
browser.visit_website('stackoverflow')
print(browser)
```
